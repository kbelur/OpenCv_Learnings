```python
#convert the image to floating point
flat_chess = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
gray_flat_chess  = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_fc_flt = np.float32(gray_flat_chess)
```

```python
#gray_fc_flt
plt.imshow(gray_fc_flt,cmap='gray')
```

```python
dst =cv2.cornerHarris(src=gray_fc_flt,blockSize=2,ksize=3,k=0.01)
```

```python
#dst
plt.imshow(dst,cmap='gray')
```

```python
dst = cv2.dilate(dst,None)
```

```python
#dst
plt.imshow(dst,cmap='gray')
```

```python
#re assigning the points above at edges to red. This is just for visualization
flat_chess[dst>0.01*dst.max()] = [255,0,0] #RGB
```

```python
plt.imshow(flat_chess)
```

```python
gray_f_rc = np.float32(gray_rc)
```

```python
dst = cv2.cornerHarris(src = gray_f_rc,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)
realchess[dst>0.01*dst.max()] = [0,255,0]
plt.imshow(realchess)
```

```python
flat_chess = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

realchess = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/real_chessboard.jpg')
realchess = cv2.cvtColor(realchess,cv2.COLOR_BGR2RGB)

gray_flat_chess  = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_real_chess  = cv2.cvtColor(realchess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,maxCorners = 64,qualityLevel = 0.01, minDistance = 10)
```

```python
#corners
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,[255,0,0],-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,80,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(realchess,(x,y),3,(0,255,0),-1)
```

```python
plt.imshow(realchess)
```
