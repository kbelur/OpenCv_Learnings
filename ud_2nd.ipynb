```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# this library will allow us to open image in an array, then we can use numpy on this
from PIL import Image
```

```python
pic = Image.open('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/00-puppy.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
pic_arr.shape
```

```python
#show the image in array using matplotlib
plt.imshow(pic_arr)
```

```python
pic_arr_red = pic_arr.copy()
```

```python
#show only red channel image 0 - red, 1 - green , 2 - blue 
plt.imshow(pic_arr_red[:,:,0])
```

```python
plt.imshow(pic_arr_red[:,:,1])
```

```python
plt.imshow(pic_arr_red[:,:,2])
```

```python
plt.imshow(pic_arr_red[:,:,0],cmap='gray') #mapping to gray scale
```

```python
plt.imshow(pic_arr_red[:,:,1],cmap='gray') #mapping to gray scale
```

```python
plt.imshow(pic_arr_red[:,:,2],cmap='gray') #mapping to gray scale
```

```python
#let me set all of the values from green channel to zero
pic_arr_red[:,:,1] = 0
```

```python
plt.imshow(pic_arr_red) #this shows only blue and red colors
```

```python
pic_arr_red[:,:,2] = 0
```

```python
plt.imshow(pic_arr_red)#this shows only red colors
```
