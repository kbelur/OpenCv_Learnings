```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/coins_on_white.jpg').astype(np.float32)/255
```

```python
display_img(img)
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,2,0,ksize=3)
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
display_img(sobely)
```

```python
kernel = np.ones(shape=(3,3),dtype=np.uint8)
```

```python
img.min()
```

```python
grad_img = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
display_img(grad_img)
```
