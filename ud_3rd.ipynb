```python
# MATPLOTLIB --> Red, Green ,Blue
# opencv     --> Blue, Green, Red

#above is the diffeence between the channels. That is why the above image is bit blue-ish
```

```python
#Hence while opening this image in opencv, we will have to transform the image from default BGR to RGB. Below is the code to do it

fix_channel_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_channel_img)
```

```python
# how to open an image and read it in gray scale
img_gray = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray)
```

```python
img_gray.shape
```

```python
#from above we can see that it is not totally into gray scale. This is due to the BGR channel in opencv2
# we can overcome this as follows
plt.imshow(img_gray,cmap='gray')
```

```python
# Re-size the image
plt.imshow(fix_channel_img)
```

```python
fix_channel_img.shape
```

```python
new_img = cv2.resize(fix_channel_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
# if you want to resize by ration
w_ratio = 0.8 # 80% of the originial
h_ratio = 0.8 #20% of the original
```

```python
new_img = cv2.resize(fix_channel_img,(0,0),fix_channel_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
# what are few ways to flip the image
flip_img = cv2.flip(fix_channel_img,-1)
fix_flip_img = cv2.cvtColor(flip_img,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(fix_flip_img)
#plt.imshow(flip_img)
```

```python
# how to save an image file
type(fix_flip_img)
```

```python
cv2.imwrite('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/00-puppy_flip.jpg',fix_flip_img)
```
