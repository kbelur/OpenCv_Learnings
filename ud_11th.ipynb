```python
import cv2

#the way we capture the webcam
cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) #it will return the resolution in floating point, so to convert it we use int
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer = cv2.VideoWriter('myvideo.mp4',cv2.VideoWriter_fourcc(*'DIVX'),20,(width,height))
while True:
    ret,frame = cap.read() #this will return a tuple, so we are storing in 2 values
    
    #perform some operation
    writer.write(frame)
    
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',gray)
    
    # org('q') ->  ordinal
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
writer.release()
cv2.destroyAllWindows()
```

```python
#Drawing on a video


import cv2

cap = cv2.VideoCapture(0)

wd = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
ht = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Top left corner
x = wd // 2  # 2 forward slash will allow x to be saved as integer
y = ht // 2

# width and height of rectangle

w = wd // 4
h = ht // 4


# Bottom right x+w, y+h


while True:
    ret, frame = cap.read()
    
    # drawing a rectangle on each video frame
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(255,0,0),thickness = 5)
    
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
# how to draw rectangle in a interactive way

import cv2

cap = cv2.VideoCapture(0)


## Create a Call back function
def draw_rect(event,x,y,flags,param):
    
    global pt1,pt2,topLeft_clicked,botRight_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        # Reset the rectangle
        if topLeft_clicked and botRight_clicked:
            pt1 = (0,0) # pointer for top left
            pt2 = (0,0) # pointer for bottom right
            topLeft_clicked = False
            botRight_clicked = False
    
        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked = True
    
        elif botRight_clicked == False:
            pt2 = (x,y)
            botRight_clicked = True

## DEFINE GLOBAL VARIABLES
pt1 = (0,0) # pointer for top left
pt2 = (0,0) # pointer for bottom right
topLeft_clicked = False
botRight_clicked = False

## Connect to CALLBACK

cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_rect)

while True:
    ret,frame = cap.read()
    
    
    # DRAW ON THE FRAME BASED OFF THE GLOBAL VARIABLE 
    if topLeft_clicked == True:
        cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),thickness=-1)
        
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame,pt1,pt2,color=(0,0,255),thickness=3)
    
    cv2.imshow('Test',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

# define the call back function
def draw_cricle(event,x,y,flags,param):
    global x1,y1,lft_but_click
    if event == cv2.EVENT_LBUTTONDOWN:
        if lft_but_click == True:
            x1 = 0
            y1 = 0
            lft_but_click = False
            
        if lft_but_click == False:
            x1,y1 = x,y
            lft_but_click = True
    

# define the variable where you have reference co-ordinates for x,y
x1 = 0
y1 = 0

lft_but_click = False

cv2.namedWindow('frame')
cv2.setMouseCallback('frame',draw_cricle)

while True:
    
    ret,frame = cap.read()

    if lft_but_click:
        cv2.circle(frame,(x1,y1),radius=30,color=(255,0,0),thickness=5)
  
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
